# Byte data_type

```python
b =[10,30,40,60,70]
type(b)
```

```python
 x=bytes(b)
type(x)
```

```python
for i in b:
    print(i)
```

```python
# bytes data is immutable(once defined cannot be changed)
# x[0] = 80
```

# bytes Array data_type

```python
y =[40,50,60,44]
by= bytearray(y)
for i in by:
    print(i)
```

```python
by[0]
```

```python
# bytearray is mutable
y[0] = 80
for i in y:
    print(i)
```
