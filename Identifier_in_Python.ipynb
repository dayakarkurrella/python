**Identifier**

A name in python program is identifier.
It can be the name of variable name, function name, module name or class name.

### Rules to define identifiers in python

Rules
1. Only Alphabet symbols are allowed (Upper and lower case)  Ex:- A-Z, a-z.
2. digits are allowed  ex:- 0-9
3. underscore is allowed ex:- _
4. $ is not allowed  ex:- ca$h =30
5. Identifier should start with character not wil number
6. Identifiers are case sensitive.
   total =10
   Total =100  both total and Total are different
7. We cannot reserve words as identifiers(35 words)
   ex: class, def, function, keyword
