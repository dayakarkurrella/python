# Type Casting (Converting one data type to other data type)

```python
a=10
type(a)
# int to float
float(a)
#int to complex
complex(a)
#int to str
str(a)
```

```python
b=12.54
```

```python
print(int(b), type(int(b)))
print(complex(b) , type(complex(b)))
print(bool(b) , type(bool(b)))
print(str(b) , type(str(b)))
```

# complex conversion

```python
com = 7+8j

# print(int(com), type(int(com)))  -->error
# print(float(com) , type(float(com)))  --error
print(bool(com) , type(bool(com)))
print(str(com) , type(str(com)))
```
